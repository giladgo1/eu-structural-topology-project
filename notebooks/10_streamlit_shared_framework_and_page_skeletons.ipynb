The original P4 versions mixed:

* challenge selection
* impact
* response building
* recovery
* interpretation

on the same screen.

This created confusion because users could not understand:

```text
What happened?
What changed?
What am I testing?
```

The final direction separates:

Recovery testing

This mirrors how users naturally think.
