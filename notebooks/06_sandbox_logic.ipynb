| Layer | Role |
|---|---|
| Inputs | User changes strategic priorities |

| Outputs | User observes directional consequences and tradeoffs |

| Repeat Process / Tracker | Learning through comparison across trials |

# Sandbox Entity Registry (Draft v1)

## Purpose

The Strategic Tradeoff Sandbox is organized into several layers.

Each layer serves a different role and should not be confused with the others.

The sandbox is designed as an educational exploration system rather than a predictive simulator.

Conceptual flow:

User Objective
→ Strategic Priorities (Inputs)
→ Structural Dimensions (Intermediate Layer)
→ Outcomes
→ Tradeoff Interpretation
→ Learning Loop
→ Documentation

---

## 1. Strategic Objectives

Strategic objectives define what the user is trying to achieve.

Objectives do not directly change the system.

Instead, they help frame interpretation and highlight relevant tradeoffs.

| Objective                 | Description                                                |
| ------------------------- | ---------------------------------------------------------- |
| Growth & Competitiveness  | Improve long-term economic performance and productivity    |
| Sustainability Transition | Improve environmental performance and transition readiness |
| Social Cohesion           | Improve social stability and reduce structural tensions    |
| Resilience & Security     | Improve shock absorption and strategic preparedness        |
| Balanced Development      | Seek balanced progress across multiple objectives          |

---

## 2. Strategic Priorities (Inputs)

Inputs represent areas of strategic emphasis that the user can adjust.

They correspond conceptually to public-investment priorities or strategic focus areas.

| Input                | Description                                                       |
| -------------------- | ----------------------------------------------------------------- |
| Education            | Human capital development and skills formation                    |
| R&D                  | Innovation and knowledge-creation investment                      |
| Sustainability       | Environmental transition and renewable-energy focus               |
| Social Protection    | Welfare, inclusion, and social support systems                    |
| Economic Development | Economic affairs, infrastructure, and productive capacity support |
| Defense              | Security, defense, and strategic preparedness                     |

---

## 3. Structural Dimensions (Intermediate Layer)

Structural dimensions represent the underlying state of a country's structural system.

They are derived from the EDA and topology work and form the bridge between priorities and outcomes.

Users observe them but do not directly manipulate them.

| Dimension                 | Description                                                            |
| ------------------------- | ---------------------------------------------------------------------- |
| Human Capital Capacity    | Skills, education, and workforce capability                            |
| Innovation Capacity       | Knowledge creation, R&D, and innovation ecosystem strength             |
| Sustainability Capacity   | Ability to support environmental transition and lower emissions        |
| Social Stability          | Social cohesion, labor stability, and inequality management            |
| Fiscal Flexibility        | Ability to absorb costs and respond without excessive fiscal stress    |
| Security Reprioritization | Strategic emphasis on defense, security, and geopolitical preparedness |
| Adaptive Transformation   | Ability to adjust and evolve under changing structural conditions      |

---

## 4. Outcomes

Outcomes represent the observable consequences of structural positioning.

They are the main results viewed by the user.

### Static Outcomes

| Outcome              | Description                                                |
| -------------------- | ---------------------------------------------------------- |
| Growth Potential     | Long-term economic growth capability                       |
| Innovation Readiness | Ability to generate and absorb innovation                  |
| Emissions Pressure   | Environmental burden and transition challenge              |
| Social Cohesion      | Degree of social stability and inclusion                   |
| Fiscal Pressure      | Degree of fiscal constraint and budget stress              |
| Structural Position  | Relative position within the European structural landscape |

### Dynamic Outcomes

| Outcome            | Description                                             |
| ------------------ | ------------------------------------------------------- |
| Shock Resilience   | Ability to absorb external shocks                       |
| Recovery Speed     | Speed of recovery following disruption                  |
| Transition Agility | Ability to adapt during structural transitions          |
| Adaptive Momentum  | Direction and strength of structural movement over time |

---

## 5. Tradeoff Interpretation Layer

Tradeoffs are the educational core of the sandbox.

The purpose is not to identify optimal strategies but to reveal tensions between competing objectives.

Examples:

* Growth vs Emissions
* Social Protection vs Fiscal Pressure
* Security vs Fiscal Flexibility
* Innovation vs Short-Term Stability
* Sustainability vs Transition Cost

---

## 6. Learning Layer

Learning occurs through repeated experimentation.

Users compare multiple strategic choices and observe how tradeoffs change.

The goal is to develop systems-thinking rather than produce predictions.

---

## 7. Documentation Layer

Exports document:

* country
* baseline
* objective
* priority choices
* observed changes
* tradeoffs identified
* interpretation summary

The export acts as a structured decision journal rather than a simulation report.

# Canonical Tradeoff Registry (Draft v1)

## Purpose

Tradeoffs are the educational core of the Strategic Tradeoff Sandbox.

The purpose of the sandbox is not to identify optimal strategies, but to help users understand tensions between competing objectives and structural pathways.

Tradeoffs should therefore be treated as first-class analytical objects rather than side effects of the model.

---

## Tradeoff Classification

### Type A — Direct / Trivial Tradeoffs

These are relationships that are relatively intuitive, mechanically linked, or commonly expected.

They remain useful but typically provide lower educational surprise.

### Type B — Structural / Non-Trivial Tradeoffs

These are tensions that emerge from complex system behavior.

These are the most valuable tradeoffs for learning, exploration, and discussion.

````python
# ============================================================
# Relationship Screening Function
# Input Spending Priorities vs Structural Dimensions
# ============================================================

input_cols = [
    "education_spending_position_eu_long_avg",
    "social_protection_spending_position_eu_long_avg",
    "defense_spending_position_eu_long_avg",
    "economic_affairs_spending_position_eu_long_avg",
    "environment_spending_position_eu_long_avg"
]

dimension_cols = [
    "dim_human_capital_capacity",
    "dim_innovation_capacity",
    "dim_sustainability_capacity",
    "dim_social_stability",
    "dim_fiscal_flexibility",
    "dim_security_reprioritization",
    "dim_adaptive_transformation"
]

relationship_rows = []

for input_col in input_cols:
    for dim_col in dimension_cols:
        corr = country_summary[input_col].corr(country_summary[dim_col])
        
        relationship_rows.append({
            "input": input_col,
            "dimension": dim_col,
            "corr": corr,
            "abs_corr": abs(corr)
        })

relationship_screening = (
    pd.DataFrame(relationship_rows)
    .sort_values("abs_corr", ascending=False)
    .reset_index(drop=True)
)

display(relationship_screening.round(3))
````

````python
# ============================================================
# Sandbox Relationship Validation Toolkit — Version 1.2
#
# Purpose:
# Validate sandbox relationships:
# spending input -> strategic dimension
#
# Outputs:
# - scatterplot by structural family
# - Pearson / Spearman / gap
# - family-level correlations
# - compact registry
#
# Notes:
# - show_plot=True for visual inspection
# - show_plot=False for rebuilding registry quickly
# - duplicate relationship names are replaced, not appended
# ============================================================

sandbox_results = []

def analyze_sandbox_relationship(
    input_col,
    dimension_col,
    relationship_name=None,
    follow_up_note="",
    mixed_signal_note="",
    show_plot=True
):
    if relationship_name is None:
        relationship_name = f"{input_col} -> {dimension_col}"

    plot_df = (
        country_summary
        .merge(
            family_metadata[["country_name", "structural_family"]],
            on="country_name",
            how="left"
        )
        .copy()
    )

    pearson_corr = plot_df[input_col].corr(plot_df[dimension_col], method="pearson")
    spearman_corr = plot_df[input_col].corr(plot_df[dimension_col], method="spearman")
    corr_gap = abs(pearson_corr - spearman_corr)
    abs_corr = abs(pearson_corr)

    family_rows = []

    for family, group in plot_df.groupby("structural_family"):
        if len(group) >= 3:
            family_pearson = group[input_col].corr(group[dimension_col], method="pearson")
            family_spearman = group[input_col].corr(group[dimension_col], method="spearman")
        else:
            family_pearson = np.nan
            family_spearman = np.nan

        family_rows.append({
            "structural_family": family,
            "n": len(group),
            "family_pearson": family_pearson,
            "family_spearman": family_spearman,
            "avg_input_position": group[input_col].mean(),
            "avg_dimension_score": group[dimension_col].mean()
        })

    family_summary = pd.DataFrame(family_rows)

    valid_family_corrs = family_summary["family_pearson"].dropna().abs()

    if len(valid_family_corrs) > 0:
        max_family_corr = valid_family_corrs.max()
        min_family_corr = valid_family_corrs.min()
        family_spread = max_family_corr - min_family_corr
    else:
        max_family_corr = np.nan
        min_family_corr = np.nan
        family_spread = np.nan

    if abs_corr >= 0.60:
        if family_spread < 0.30:
            screening_classification = "Universal Candidate"
        else:
            screening_classification = "Mixed Candidate"
    elif abs_corr < 0.30 and max_family_corr >= 0.60:
        screening_classification = "Conditional Candidate"
    elif abs_corr >= 0.30:
        screening_classification = "Medium / Needs Inspection"
    else:
        screening_classification = "Weak / Unsupported"

    result_row = {
        "relationship": relationship_name,
        "input_col": input_col,
        "dimension_col": dimension_col,
        "pearson": pearson_corr,
        "spearman": spearman_corr,
        "corr_gap": corr_gap,
        "abs_pearson": abs_corr,
        "max_family_corr": max_family_corr,
        "min_family_corr": min_family_corr,
        "family_spread": family_spread,
        "screening_classification": screening_classification,
        "mixed_signal_note": mixed_signal_note,
        "follow_up_note": follow_up_note
    }

    # Replace duplicate relationship if it already exists
    global sandbox_results
    sandbox_results = [
        row for row in sandbox_results
        if row["relationship"] != relationship_name
    ]
    sandbox_results.append(result_row)

    sandbox_registry = (
        pd.DataFrame(sandbox_results)
        .sort_values("abs_pearson", ascending=False)
        .reset_index(drop=True)
    )

    compact_registry = sandbox_registry[
        [
            "relationship",
            "pearson",
            "spearman",
            "corr_gap",
            "family_spread",
            "screening_classification",
            "mixed_signal_note",
            "follow_up_note"
        ]
    ]

    if show_plot:
        plt.figure(figsize=(10, 7))

        sns.scatterplot(
            data=plot_df,
            x=input_col,
            y=dimension_col,
            hue="structural_family",
            s=120
        )

        archetypes = plot_df[plot_df["is_archetype_or_reference"] == True]

        for _, row in archetypes.iterrows():
            plt.annotate(
                row["country_name"],
                (row[input_col], row[dimension_col]),
                fontsize=8
            )

        plt.title(
            f"{relationship_name}\n"
            f"Pearson = {pearson_corr:.2f} | "
            f"Spearman = {spearman_corr:.2f} | "
            f"Gap = {corr_gap:.2f} | "
            f"n = {len(plot_df)}"
        )

        plt.xlabel(input_col)
        plt.ylabel(dimension_col)

        plt.tight_layout()
        plt.show()

    print("============================================================")
    print("RELATIONSHIP VALIDATION")
    print("============================================================")
    print(f"Relationship: {relationship_name}")
    print(f"Pearson:      {pearson_corr:.3f}")
    print(f"Spearman:     {spearman_corr:.3f}")
    print(f"Corr gap:     {corr_gap:.3f}")
    print(f"Classification: {screening_classification}")
    print(f"Mixed signal: {mixed_signal_note}")
    print(f"Follow-up: {follow_up_note}")

    print("\nFamily-level summary:")
    display(family_summary.round(3))

    print("\nCompact sandbox registry:")
    display(compact_registry.round(3))

    return family_summary, sandbox_registry, compact_registry
````

````python

# ============================================================
# Task 3B — Intermediate → Output Relationship Mapping
# ============================================================

dimension_cols = [
    "dim_human_capital_capacity",
    "dim_innovation_capacity",
    "dim_sustainability_capacity",
    "dim_social_stability",
    "dim_fiscal_flexibility",
    "dim_security_reprioritization",
    "dim_adaptive_transformation"
]

dimension_corr = country_summary[dimension_cols].corr()

display(
    dimension_corr
    .round(3)
    .style.background_gradient(cmap="RdYlGn")
)
````

````python
# ============================================================
# Task 3B — Dimension Relationship Validation Toolkit v1.0
#
# Purpose:
# Validate dimension → dimension / intermediate → output links
# using the same workflow as Task 3A.
#
# Outputs:
# - scatterplot by structural family
# - Pearson / Spearman / gap
# - family-level correlations
# - family spread
# - screening classification
# - follow-up note
# - registry update
# ============================================================

dimension_results = []

def analyze_dimension_relationship_full(
    x_col,
    y_col,
    relationship_name=None,
    follow_up_note="",
    show_plot=True
):
    if relationship_name is None:
        relationship_name = f"{x_col} ↔ {y_col}"

    plot_df = (
        country_summary
        .merge(
            family_metadata[
                ["country_name", "structural_family"]
            ],
            on="country_name",
            how="left"
        )
        .copy()
    )

    # --------------------------------------------------------
    # Overall correlations
    # --------------------------------------------------------
    pearson_corr = plot_df[x_col].corr(
        plot_df[y_col],
        method="pearson"
    )

    spearman_corr = plot_df[x_col].corr(
        plot_df[y_col],
        method="spearman"
    )

    corr_gap = abs(pearson_corr - spearman_corr)
    abs_corr = abs(pearson_corr)

    # --------------------------------------------------------
    # Family-level correlations
    # --------------------------------------------------------
    family_rows = []

    for family, group in plot_df.groupby("structural_family"):

        if len(group) >= 3:
            family_pearson = group[x_col].corr(
                group[y_col],
                method="pearson"
            )

            family_spearman = group[x_col].corr(
                group[y_col],
                method="spearman"
            )
        else:
            family_pearson = np.nan
            family_spearman = np.nan

        family_rows.append({
            "structural_family": family,
            "n": len(group),
            "family_pearson": family_pearson,
            "family_spearman": family_spearman,
            "avg_x_score": group[x_col].mean(),
            "avg_y_score": group[y_col].mean()
        })

    family_summary = pd.DataFrame(family_rows)

    valid_family_corrs = (
        family_summary["family_pearson"]
        .dropna()
        .abs()
    )

    if len(valid_family_corrs) > 0:
        max_family_corr = valid_family_corrs.max()
        min_family_corr = valid_family_corrs.min()
        family_spread = max_family_corr - min_family_corr
    else:
        max_family_corr = np.nan
        min_family_corr = np.nan
        family_spread = np.nan

    # --------------------------------------------------------
    # Automatic screening classification
    # --------------------------------------------------------
    if abs_corr >= 0.60:

        if family_spread < 0.30:
            screening_classification = "Strong / Universal Candidate"
        else:
            screening_classification = "Strong / Mixed Candidate"

    elif abs_corr >= 0.30:

        if max_family_corr >= 0.60:
            screening_classification = "Medium / Family-Dependent Candidate"
        else:
            screening_classification = "Medium / Needs Inspection"

    elif abs_corr < 0.30 and max_family_corr >= 0.60:

        screening_classification = "Conditional Candidate"

    else:

        screening_classification = "Weak / Unsupported"

    # --------------------------------------------------------
    # Save result into registry
    # --------------------------------------------------------
    result_row = {
        "relationship": relationship_name,
        "x_col": x_col,
        "y_col": y_col,
        "pearson": pearson_corr,
        "spearman": spearman_corr,
        "corr_gap": corr_gap,
        "abs_pearson": abs_corr,
        "max_family_corr": max_family_corr,
        "min_family_corr": min_family_corr,
        "family_spread": family_spread,
        "screening_classification": screening_classification,
        "follow_up_note": follow_up_note
    }

    global dimension_results
    dimension_results = [
        row for row in dimension_results
        if row["relationship"] != relationship_name
    ]

    dimension_results.append(result_row)

    dimension_registry = (
        pd.DataFrame(dimension_results)
        .sort_values("abs_pearson", ascending=False)
        .reset_index(drop=True)
    )

    compact_dimension_registry = dimension_registry[
        [
            "relationship",
            "pearson",
            "spearman",
            "corr_gap",
            "family_spread",
            "screening_classification",
            "follow_up_note"
        ]
    ]

    # --------------------------------------------------------
    # Scatterplot
    # --------------------------------------------------------
    if show_plot:

        plt.figure(figsize=(10, 7))

        sns.scatterplot(
            data=plot_df,
            x=x_col,
            y=y_col,
            hue="structural_family",
            s=120
        )

        archetypes = plot_df[
            plot_df["is_archetype_or_reference"] == True
        ]

        for _, row in archetypes.iterrows():
            plt.annotate(
                row["country_name"],
                (row[x_col], row[y_col]),
                fontsize=8
            )

        plt.title(
            f"{relationship_name}\n"
            f"Pearson = {pearson_corr:.2f} | "
            f"Spearman = {spearman_corr:.2f} | "
            f"Gap = {corr_gap:.2f} | "
            f"n = {len(plot_df)}"
        )

        plt.xlabel(x_col)
        plt.ylabel(y_col)

        plt.tight_layout()
        plt.show()

    # --------------------------------------------------------
    # Printed summary
    # --------------------------------------------------------
    print("============================================================")
    print("DIMENSION RELATIONSHIP VALIDATION")
    print("============================================================")
    print(f"Relationship: {relationship_name}")
    print(f"Pearson:      {pearson_corr:.3f}")
    print(f"Spearman:     {spearman_corr:.3f}")
    print(f"Corr gap:     {corr_gap:.3f}")
    print(f"Classification: {screening_classification}")
    print(f"Follow-up: {follow_up_note}")

    print("\nFamily-level summary:")
    display(family_summary.round(3))

    print("\nCompact dimension registry:")
    display(compact_dimension_registry.round(3))

    return family_summary, dimension_registry, compact_dimension_registry
````

````python
# ============================================================
# Task 3C — Sandbox Response Matrix v1
#
# Evidence weights:
# A = 85 strong response
# B = 65 moderate response
# C = 35 conditional / family-dependent response
# D = 10 weak exploratory signal
#
# These are NOT causal coefficients.
# They are confidence-adjusted sandbox response strengths.
# ============================================================

evidence_weight_map = {
    "A": 85,
    "B": 65,
    "C": 35,
    "D": 10
}

sandbox_response_rows = [

    # Investment → Dimension
    ["Investment", "Education", "Human Capital", "A"],
    ["Investment", "Defense", "Security", "A"],
    ["Investment", "Social Protection", "Fiscal Flexibility", "A"],

    ["Investment", "Social Protection", "Social Stability", "B"],
    ["Investment", "Education", "Innovation", "B"],
    ["Investment", "Social Protection", "Innovation", "B"],

    ["Investment", "Environment", "Sustainability", "C"],
    ["Investment", "Social Protection", "Adaptive Transformation", "C"],

    ["Investment", "Education", "Social Stability", "D"],
    ["Investment", "Education", "Sustainability", "D"],
    ["Investment", "Defense", "Adaptive Transformation", "D"],
    ["Investment", "Economic Affairs", "Fiscal Flexibility", "D"],
    ["Investment", "Education", "Fiscal Flexibility", "D"],
    ["Investment", "Economic Affairs", "Innovation", "D"],
    ["Investment", "Economic Affairs", "Sustainability", "D"],
    ["Investment", "Economic Affairs", "Adaptive Transformation", "D"],
    ["Investment", "Defense", "Fiscal Flexibility", "D"],
    ["Investment", "Environment", "Adaptive Transformation", "D"],

    # Dimension → Dimension
    ["Dimension", "Innovation", "Social Stability", "A"],

    ["Dimension", "Social Stability", "Security", "B"],

    ["Dimension", "Human Capital", "Innovation", "C"],
    ["Dimension", "Fiscal Flexibility", "Adaptive Transformation", "C"],
    ["Dimension", "Sustainability", "Fiscal Flexibility", "C"],

    ["Dimension", "Innovation", "Security", "D"],
    ["Dimension", "Human Capital", "Social Stability", "D"],
    ["Dimension", "Social Stability", "Fiscal Flexibility", "D"],
    ["Dimension", "Sustainability", "Security", "D"],
]

sandbox_response_matrix_v1 = pd.DataFrame(
    sandbox_response_rows,
    columns=[
        "layer",
        "source",
        "target",
        "evidence_level"
    ]
)

sandbox_response_matrix_v1["base_weight"] = (
    sandbox_response_matrix_v1["evidence_level"]
    .map(evidence_weight_map)
)

display(sandbox_response_matrix_v1)
````

````python
# ============================================================
# Task 3C-B — Transparent Family Response Modifiers v2
#
# Logic:
# family_modifier is derived from family Pearson behavior
# relative to the overall relationship direction.
#
# Not regression.
# Not causal.
# Transparent correlation-based adjustment.
# ============================================================

family_modifier_source = pd.DataFrame([

    # layer, source, target, overall_pearson, family, family_pearson

    ["Investment", "Environment", "Sustainability", 0.118, "Adaptive / Peripheral Systems", 0.317],
    ["Investment", "Environment", "Sustainability", 0.118, "Industrial / Transition Systems", 0.729],
    ["Investment", "Environment", "Sustainability", 0.118, "Innovation-Core Systems", -0.246],

    ["Investment", "Social Protection", "Adaptive Transformation", 0.345, "Adaptive / Peripheral Systems", 0.715],
    ["Investment", "Social Protection", "Adaptive Transformation", 0.345, "Industrial / Transition Systems", 0.827],
    ["Investment", "Social Protection", "Adaptive Transformation", 0.345, "Innovation-Core Systems", -0.547],

    ["Dimension", "Human Capital", "Innovation", 0.406, "Adaptive / Peripheral Systems", 0.075],
    ["Dimension", "Human Capital", "Innovation", 0.406, "Industrial / Transition Systems", 0.102],
    ["Dimension", "Human Capital", "Innovation", 0.406, "Innovation-Core Systems", 0.937],

    ["Dimension", "Fiscal Flexibility", "Adaptive Transformation", -0.370, "Adaptive / Peripheral Systems", -0.718],
    ["Dimension", "Fiscal Flexibility", "Adaptive Transformation", -0.370, "Industrial / Transition Systems", -0.363],
    ["Dimension", "Fiscal Flexibility", "Adaptive Transformation", -0.370, "Innovation-Core Systems", 0.814],

    ["Dimension", "Sustainability", "Fiscal Flexibility", -0.308, "Adaptive / Peripheral Systems", -0.220],
    ["Dimension", "Sustainability", "Fiscal Flexibility", -0.308, "Industrial / Transition Systems", -0.768],
    ["Dimension", "Sustainability", "Fiscal Flexibility", -0.308, "Innovation-Core Systems", 0.459],

], columns=[
    "layer",
    "source",
    "target",
    "overall_pearson",
    "structural_family",
    "family_pearson"
])

# Direction agreement:
# +1 = same direction as overall relationship
# -1 = opposite direction
family_modifier_source["direction_agreement"] = (
    family_modifier_source["overall_pearson"]
    * family_modifier_source["family_pearson"]
).apply(lambda x: 1 if x >= 0 else -1)

# Strength difference:
# family relationship strength minus overall strength
family_modifier_source["strength_delta"] = (
    family_modifier_source["family_pearson"].abs()
    - family_modifier_source["overall_pearson"].abs()
)

# Convert to modifier:
# every 0.10 correlation difference ≈ 5 weight points
family_modifier_source["raw_modifier"] = (
    family_modifier_source["strength_delta"] * 50
)

# Reverse sign if family relationship goes opposite direction
family_modifier_source["family_modifier"] = (
    family_modifier_source["raw_modifier"]
    * family_modifier_source["direction_agreement"]
)

# Keep modifiers bounded
family_modifier_source["family_modifier"] = (
    family_modifier_source["family_modifier"]
    .round()
    .clip(lower=-40, upper=40)
)

family_response_modifier_v1 = family_modifier_source[
    [
        "layer",
        "source",
        "target",
        "structural_family",
        "overall_pearson",
        "family_pearson",
        "direction_agreement",
        "strength_delta",
        "family_modifier"
    ]
]

display(family_response_modifier_v1)
````

````python
sandbox_response_family_weights_v1 = (
    sandbox_response_matrix_v1
    .merge(
        family_response_modifier_v1[
            [
                "layer",
                "source",
                "target",
                "structural_family",
                "family_modifier"
            ]
        ],
        on=["layer", "source", "target"],
        how="left"
    )
)

sandbox_response_family_weights_v1["family_modifier"] = (
    sandbox_response_family_weights_v1["family_modifier"]
    .fillna(0)
)

sandbox_response_family_weights_v1["effective_weight"] = (
    sandbox_response_family_weights_v1["base_weight"]
    + sandbox_response_family_weights_v1["family_modifier"]
)

sandbox_response_family_weights_v1["effective_weight"] = (
    sandbox_response_family_weights_v1["effective_weight"]
    .clip(lower=-85, upper=85)
)

display(sandbox_response_family_weights_v1)
````

````python
# ============================================================
# Manual Direction Overrides
# Based on validated family reversals
# ============================================================

sandbox_response_family_weights_v1["manual_direction_override"] = False

# ------------------------------------------------------------
# Social Protection → Adaptive Transformation
# Innovation-Core reversal
# ------------------------------------------------------------

mask = (
    (sandbox_response_family_weights_v1["source"] == "Social Protection") &
    (sandbox_response_family_weights_v1["target"] == "Adaptive Transformation") &
    (sandbox_response_family_weights_v1["structural_family"] == "Innovation-Core Systems")
)

sandbox_response_family_weights_v1.loc[
    mask,
    "effective_weight"
] = -25

sandbox_response_family_weights_v1.loc[
    mask,
    "manual_direction_override"
] = True

# ------------------------------------------------------------
# Fiscal Flexibility → Adaptive Transformation
# Full family reversal logic
# ------------------------------------------------------------

override_values = {
    "Adaptive / Peripheral Systems": -52,
    "Industrial / Transition Systems": -35,
    "Innovation-Core Systems": 42
}

for family, weight in override_values.items():

    mask = (
        (sandbox_response_family_weights_v1["source"] == "Fiscal Flexibility") &
        (sandbox_response_family_weights_v1["target"] == "Adaptive Transformation") &
        (sandbox_response_family_weights_v1["structural_family"] == family)
    )

    sandbox_response_family_weights_v1.loc[
        mask,
        "effective_weight"
    ] = weight

    sandbox_response_family_weights_v1.loc[
        mask,
        "manual_direction_override"
    ] = True

# ------------------------------------------------------------
# Sustainability → Fiscal Flexibility
# Full family reversal logic
# ------------------------------------------------------------

override_values = {
    "Adaptive / Peripheral Systems": -31,
    "Industrial / Transition Systems": -58,
    "Innovation-Core Systems": 27
}

for family, weight in override_values.items():

    mask = (
        (sandbox_response_family_weights_v1["source"] == "Sustainability") &
        (sandbox_response_family_weights_v1["target"] == "Fiscal Flexibility") &
        (sandbox_response_family_weights_v1["structural_family"] == family)
    )

    sandbox_response_family_weights_v1.loc[
        mask,
        "effective_weight"
    ] = weight

    sandbox_response_family_weights_v1.loc[
        mask,
        "manual_direction_override"
    ] = True

# ------------------------------------------------------------
# Review overrides
# ------------------------------------------------------------

display(
    sandbox_response_family_weights_v1[
        sandbox_response_family_weights_v1["manual_direction_override"] == True
    ]
)

print("Manual reversal corrections applied.")
````
