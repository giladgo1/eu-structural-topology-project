## Step 2 — Create App-Ready Country Profile Dictionary

Goal:

Convert the raw Germany row into a clean dictionary that the Streamlit app can use.

Why this matters:

Streamlit pages should not repeatedly filter raw data inside every visual block.
Instead, each selected country should be converted into a simple profile object containing:

- identity
- archetype
- family
- dimensions
- strongest dimension
- weakest dimension

This object will later power:

- context ribbon
- country story card
- dimension snapshot
- strengths and constraints
- reflection page

`````python
DIMENSION_LABELS = {
    "dim_sustainability_capacity": "Sustainability Capacity",
    "dim_human_capital_capacity": "Human Capital Capacity",
    "dim_innovation_capacity": "Innovation Capacity",
    "dim_social_stability": "Social Stability",
    "dim_fiscal_flexibility": "Fiscal Flexibility",
    "dim_security_reprioritization": "Security Reprioritization",
    "dim_adaptive_transformation": "Adaptive Transformation",
}

dimension_cols = list(DIMENSION_LABELS.keys())

def build_country_profile(country_name, profiles_df, families_df):
    """
    Build a clean app-ready country profile object.
    This will later become reusable Streamlit logic.
    """
    profile_row = profiles_df.loc[
        profiles_df["country_name"] == country_name
    ].iloc[0]

    family_row = families_df.loc[
        families_df["country_name"] == country_name
    ].iloc[0]

    dimensions = {
        DIMENSION_LABELS[col]: float(profile_row[col])
        for col in dimension_cols
    }

    strongest_dimension = max(dimensions, key=dimensions.get)
    weakest_dimension = min(dimensions, key=dimensions.get)

    country_profile = {
        "country": profile_row["country_name"],
        "country_code": profile_row["country"],
        "archetype": profile_row["archetype_label"],
        "archetype_type": profile_row["archetype_type"],
        "family": family_row["structural_family"],
        "subfamily": family_row["structural_subfamily"],
        "family_anchor": family_row["family_anchor_archetype"],
        "family_color": family_row["family_color"],
        "dimensions": dimensions,
        "strongest_dimension": strongest_dimension,
        "strongest_value": dimensions[strongest_dimension],
        "weakest_dimension": weakest_dimension,
        "weakest_value": dimensions[weakest_dimension],
    }

    return country_profile


germany = build_country_profile(
    country_name="Germany",
    profiles_df=profiles,
    families_df=families
)

germany
`````

## Step 6 — Build Yearly Dimension Trajectories

Goal:
Create yearly dimension scores for Germany so Page 1 can answer:

"How did Germany become this way?"

Decision:
Use `dashboard1_country_year_full_norm.csv` and recalculate dimensions per year from EU-long z-score KPI columns.

This keeps the evolution section aligned with the approved P1 architecture.

`````python
country_year = dfs["country_year"].copy()

def add_yearly_dimensions(df):
    """
    Add yearly structural dimension scores using EU-long z-score KPI columns.
    These formulas mirror the static dimension logic used in the project.
    """

    df = df.copy()

    df["dim_sustainability_capacity"] = (
        0.40 * df["renewables_eu_long_zscore"]
        + 0.30 * df["environment_spending_eu_long_zscore"]
        - 0.30 * df["emissions_eu_long_zscore"]
    )

    df["dim_human_capital_capacity"] = (
        0.60 * df["education_eu_long_zscore"]
        + 0.40 * df["education_spending_eu_long_zscore"]
    )

    df["dim_innovation_capacity"] = (
        0.45 * df["rnd_eu_long_zscore"]
        + 0.35 * df["ict_specialists_eu_long_zscore"]
        + 0.20 * df["economic_affairs_spending_eu_long_zscore"]
    )

    df["dim_social_stability"] = (
        0.30 * df["social_protection_spending_eu_long_zscore"]
        + 0.25 * df["health_spending_eu_long_zscore"]
        - 0.25 * df["unemployment_eu_long_zscore"]
        - 0.20 * df["gini_eu_long_zscore"]
    )

    df["dim_fiscal_flexibility"] = (
        -0.60 * df["debt_eu_long_zscore"]
        -0.40 * df["total_gov_expenditure_eu_long_zscore"]
    )

    df["dim_security_reprioritization"] = (
        0.80 * df["defense_spending_eu_long_zscore"]
        + 0.20 * df["fuel_energy_spending_eu_long_zscore"]
    )

    df["dim_adaptive_transformation"] = (
        0.50 * df["dim_innovation_capacity"]
        + 0.50 * df["dim_sustainability_capacity"]
    )

    return df


country_year_dims = add_yearly_dimensions(country_year)

germany_yearly_dims = country_year_dims[
    country_year_dims["country_name"] == "Germany"
][
    ["country_name", "year"] + list(DIMENSION_LABELS.keys())
]

germany_yearly_dims
`````
