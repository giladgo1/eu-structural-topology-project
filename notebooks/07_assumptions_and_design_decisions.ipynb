# 07 — Assumptions, Design Decisions & Architecture Validation

## Purpose

This notebook serves as the architectural memory of the project.

The project evolved through multiple stages:

```text
EDA
→ Structural Dimensions
→ Archetypes
→ Tradeoff Space
→ Sandbox Logic
→ Shock Engine
→ Interpretation Layer
```

Many important decisions emerged through iterative analysis and discussion rather than from a single statistical model.

The purpose of this notebook is to:

* document assumptions,
* record design decisions,
* identify known limitations,
* preserve rejected alternatives,
* support future development,
* improve transparency,
* and prevent architectural drift.

---

# Part 1 — Project Philosophy

## Why This Project Exists

The project is not attempting to answer:

```text
What policy should Europe adopt?
```

Instead it asks:

```text
How can citizens, students, researchers and decision-makers
explore strategic tradeoffs under uncertainty?
```

The project therefore prioritizes:

* transparency,
* interpretability,
* exploration,
* systems thinking,
* decision awareness.

```text
Educational Sandbox
Decision Support Prototype
Strategic Exploration Tool
```

---

## What This Project Is Not

# Part 2 — Data Assumptions

## A01 — Structural Dimensions Are Constructed

### Challenge

Dimensions such as:

```text
Human Capital
Innovation
Adaptive Transformation
```

are not directly observed variables.

### Decision

Dimensions are intentionally constructed from multiple indicators.

### Reason

Strategic systems cannot be represented by single KPIs.

### Limitation

Different researchers might construct dimensions differently.

### Status

Accepted MVP Assumption.

---

## A02 — Country Archetypes Represent Pathways

### Challenge

Countries rarely fit perfectly into one category.

### Decision

Archetypes represent strategic pathways rather than fixed categories.

### Limitation

Many countries behave as hybrids.

### Status

Accepted MVP Assumption.

---

## A03 — Structural Families Are Useful

### Evidence

Clustering and pathway analysis showed recurring patterns.

### Decision

Families remain part of the architecture.

### Limitation

Family boundaries are fuzzy.

### Future

Confidence-aware family assignment.

# A19 — Dashboard Separation Principle

## Challenge

Should Dashboard 1 and Dashboard 2 be merged into a single experience?

## Evidence

Architecture reviews repeatedly revealed two fundamentally different user questions:

```text
Page 1:
What structural pathways characterize Europe?

Page 2:
What happens if I explore different strategic priorities?
```

Exploration and experimentation serve different educational purposes.

## MVP Decision

Maintain two separate pages:

```text
Page 1 explains.

Page 2 teaches.
```

## Reason

Separating observation from experimentation reduces complexity and improves user understanding.

## Known Limitation

Users may initially perceive the two pages as disconnected.

## Future Direction

Shared state and guided transitions between pages.

## Status

Accepted Architecture Freeze

---

# A23 — Family-Dependent Relationships Are Allowed

## Challenge

Some relationships appear weak at the European level but strong inside structural families.

## Evidence

Examples discovered during validation:

```text
Environment Spending
→ Sustainability

Europe-wide:
Weak

Industrial Family:
Strong

Innovation-Core:
Different direction
```

## MVP Decision

Family-specific effects are permitted.

## Reason

Observed structural heterogeneity is itself a meaningful finding.

## Known Limitation

Family sample sizes remain small.

## Future Direction

Sub-family and confidence-aware modifiers.

## Status

Accepted Architecture Decision

## Known Limitation

Accepted MVP Assumption

# A25 — D-Level Relationships Remain in MVP

## MVP Decision

## Status

Accepted MVP Assumption

# A26 — No Recursive Ecosystem Propagation

## Challenge

Dimension ecosystems suggest that dimensions influence each other.

## Rejected Alternative

```text
Investment
→ Dimension
→ Dimension
→ Dimension
→ Output
```

## MVP Decision

Dimension ecosystems remain interpretive only.

No recursive propagation occurs inside the sandbox engine.

## Reason

Avoids:

```text
Double counting
Feedback loops
Hidden amplification
Runaway effects
```

## Known Limitation

Some indirect relationships are intentionally ignored.

## Status

Critical Architecture Freeze
